```python
m = int(input("Enter the three digit number"))
a = m // 100
b = (m % 100) // 10
c = m % 10
if a!=b and b!=c and a!=c:
  print("All digits are distinct")
else:
  print("Digits are not distinct ")
```

```python
n = int(input("Enter a 3-digit number: "))

a = n // 100
b = (n // 10) % 10
c = n % 10

if b > a and b > c:
    print("Middle digit is largest")
elif b < a and b < c:
    print("Middle digit is smallest")
else:
    print("Middle digit is neither largest nor smallest")
```
